##### Changing the values of Propertycount, Distance, Bedroom2, Bathroom

```python
print("Missing values before imputation \n",data1[['Propertycount','Distance','Bedroom2','Bathroom']].isnull().sum())
```

```python
data1['Propertycount'].fillna(0, inplace = True)
data1['Distance'].fillna(0, inplace = True)
data1['Bedroom2'].fillna(0, inplace = True)
data1['Bathroom'].fillna(0, inplace = True)
```

```python
print("After imputation : \n", data1[['Propertycount','Distance','Bedroom2','Bathroom']].isnull().sum())
```

##### Changing the value of Land Size and Building Area

```python
print("Missing values before imputation : \n",data1[['Landsize','BuildingArea']].isnull().sum())
```

```python
data1['Landsize'].fillna(data1['Landsize'].mean(), inplace = True)
data1['BuildingArea'].fillna(data1['BuildingArea'].mean(), inplace = True)
```

```python
print("Missing value after imputation : \n",data1[['Landsize','BuildingArea']].isnull().sum())
```
