```python
x = data1.drop(['salary($)'],axis=1)
y = data1['salary($)']
```

```python
y
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = 69)
```

```python
print('Missing values before imputation \n', train_x.isnull().sum())
```

```python
train_x.fillna(train_x.mode().loc[0], inplace = True)
```

```python
print('Missing values after imputation \n', train_x.isnull().sum())
```

```python
train_x.reset_index(inplace=True,drop=True)
test_x.reset_index(inplace=True,drop=True)
```

```python
train_x1 = train_x
test_x1 = test_x
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler = StandardScaler()
scaler.fit(train_x1)
train_x1 = pd.DataFrame(scaler.transform(train_x1), columns = train_x1.columns)
test_x1 = pd.DataFrame(scaler.transform(test_x1), columns = test_x1.columns)
```

```python
import xgboost as xgb
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

model1 = LinearRegression()
model2 = DecisionTreeRegressor(random_state = 69)
model3 = RandomForestRegressor(random_state = 69)
model4 = KNeighborsRegressor()
model5 = xgb.XGBRegressor()
model6 = SVR()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
model1.fit(train_x1,train_y)
pred1 = model1.predict(test_x1)

mae1 = mean_absolute_error(test_y, pred1)

print(mae1)
```

```python
slope = model1.coef_[0]
intercept = model1.intercept_

print("Slope:", slope)
print("Intercept:", intercept)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

# Create and fit a linear regression model
model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# Extract slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Generate x values for the regression line
x_regression = np.linspace(min(x), max(x), 100)
y_regression = slope * x_regression + intercept

# Create a scatter plot with the regression line
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, color="blue", label="Data Points")
sns.lineplot(x=x_regression, y=y_regression, color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
